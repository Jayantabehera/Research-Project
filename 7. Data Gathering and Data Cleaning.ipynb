# Read csv files

```python
#read day 1 data
df0 = pd.read_csv("D:\\Research Project\\Datasets\\Flood intensity\\data url\\Day 1 file\day 1 4924 not null rows processed.csv")
print(len(df0)) ####4924
df0.head()
```

```python
#drop unnecessary columns
df0.drop(df0.columns[[0,1,3,5,11,16,17,18]], axis=1, inplace=True)
df0
```

```python
#convert string fields into date field
#df1['Ended'] = df1['Ended'].astype(date)
df0['Ended'] =  pd.to_datetime(df0['Ended'])
df0['Began'] = df0['Began'].astype('datetime64[ns]')
#Get the flood duration
df0['Duration'] = (df0['Ended']  - df0['Began']).astype(str)
df0.dtypes
```

```python
#Calculate Flood Duration
for i, row in df0.iterrows():
    Duration = df0.loc[i,"Duration"]
    df0.loc[i,"Duration"]= int(Duration[:2])

    
```

```python
#Calculate Area Affected oer Day
df0['Area_Affected_per_day'] = round((df0['Area']  / df0['Duration']),2)

df0.dtypes
```

```python
#data cleaning - removed Ã‚Â F values from Temperature

df0['Temperature']=df0['Temperature'].replace('Ã‚','',regex=True)
df0['Temperature']=df0['Temperature'].replace('F','',regex=True)


#data cleaning - removed Ã‚Â F values from Dew Point
df0['Dew Point'] = df0['Dew Point'].str.translate(str.maketrans({'Ã‚':''}))
df0['Dew Point'] = df0['Dew Point'].str.translate(str.maketrans({'F':''}))

#data cleaning - removed Ã‚Â % values from Humidity
df0['Humidity'] = df0['Humidity'].str.translate(str.maketrans({'Ã‚':''}))
df0['Humidity'] = df0['Humidity'].str.translate(str.maketrans({'%':''}))

#data cleaning - removed Ã‚Â mph values from Wind Speed
df0['Wind Speed'] = df0['Wind Speed'].str.translate(str.maketrans({'Ã‚':''}))
df0['Wind Speed'] = df0['Wind Speed'].str.translate(str.maketrans({'m':''}))
df0['Wind Speed'] = df0['Wind Speed'].str.translate(str.maketrans({'p':''}))
df0['Wind Speed'] = df0['Wind Speed'].str.translate(str.maketrans({'h':''}))

#data cleaning - removed Ã‚Â mph values from Wind Gust
df0['Wind Gust'] = df0['Wind Gust'].str.translate(str.maketrans({'Ã‚':''}))
df0['Wind Gust'] = df0['Wind Gust'].str.translate(str.maketrans({'m':''}))
df0['Wind Gust'] = df0['Wind Gust'].str.translate(str.maketrans({'p':''}))
df0['Wind Gust'] = df0['Wind Gust'].str.translate(str.maketrans({'h':''}))

#data cleaning - removed Ã‚Â in values from Pressure
df0['Pressure'] = df0['Pressure'].str.translate(str.maketrans({'Ã‚':''}))
df0['Pressure'] = df0['Pressure'].str.translate(str.maketrans({'i':''}))
df0['Pressure'] = df0['Pressure'].str.translate(str.maketrans({'n':''}))

#data cleaning - removed Ã‚Â in values from Precipitation
df0['Precipitation']= df0['Precipitation'].str.translate(str.maketrans({'Ã‚':''}))
df0['Precipitation']= df0['Precipitation'].str.translate(str.maketrans({'i':''}))
df0['Precipitation']= df0['Precipitation'].str.translate(str.maketrans({'n':''}))
```
